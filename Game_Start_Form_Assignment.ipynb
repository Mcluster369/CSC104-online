```python
from IPython.display import HTML

HTML('''
<!DOCTYPE html>
<html>
<head>
  <style>
    body {
      font-family: Arial, sans-serif;
      padding: 20px;
    }
    label {
      display: block;
      margin-top: 10px;
    }
    input, button {
      margin-top: 5px;
      padding: 6px;
      font-size: 14px;
    }
    #output {
      margin-top: 20px;
      background: #f4f4f4;
      padding: 10px;
      border-radius: 6px;
    }
  </style>
</head>
<body>
  <h2>Game Start Form</h2>
  <form id="gameForm">
    <label>Username:
      <input type="text" id="username" required>
    </label>
    <label>Weapons:
      <input type="text" id="weapons" required>
    </label>
    <label>Health/Damage:
      <input type="number" id="health" required>
    </label>
    <label>Point Total:
      <input type="number" id="points" required>
    </label>
    <button type="button" onclick="submitForm()">Start Game</button>
  </form>

  <div id="output"></div>

  <script>
    function submitForm() {
      var username = document.getElementById("username").value;
      var weapons = document.getElementById("weapons").value;
      var health = document.getElementById("health").value;
      var points = document.getElementById("points").value;

      var output = `
        <h3>Game Info</h3>
        <p><strong>User Name:</strong> ${username}</p>
        <p><strong>Weapons:</strong> ${weapons}</p>
        <p><strong>Health/Damage:</strong> ${health}</p>
        <p><strong>Point Total:</strong> ${points}</p>
      `;

      document.getElementById("output").innerHTML = output;
    }
  </script>
</body>
</html>
''')
```
